```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure 

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8) #Adjust the configuration of plots


df= pd.read_csv(r'C:\Users\DELL\Downloads\movies.csv')
```

```python
#Looking for Missing data

for col in df.columns :
    pct_missing = np.mean(df[col].isnull())
    print('{} = {}%'.format(col, pct_missing))
```

```python
df.dtypes
```

```python
df['budget'] = df['budget'].fillna(0)
df['gross'] =  df['gross'].fillna(0)
```

```python
 df['budget'] =  df['budget'].astype('int64')
 df['gross'] =  df['gross'].astype('int64')
```

```python
df.head()
```

```python
#Creating correct the year column

df['year_correct'] = df['released'].str.extract(r'(\d{4})')
```

```python
#Correlation
df.corr(numeric_only = True)
```

```python
#Three main methods
correlation_matrix = df.corr(method='pearson' , numeric_only=True)
plt.title('Correlation Matric for Numeric Funtion')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
df_numerized =df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object') :
        df_numerized[col_name]   = df_numerized[col_name].astype('category')
        df_numerized[col_name]   = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df.corr(method='pearson' , numeric_only=True)

plt.title('Correlation Matric for Numeric Funtion')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]

high_corr
```
